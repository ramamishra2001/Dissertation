```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science','notebook','grid'])
import pandas as pd
```

```python
data=pd.read_csv('history.csv')
data1=data.iloc[:,:5]
data2=data.iloc[:,5:-1]
data2
data1.columns
```

```python
labels=['Loss', 'Accuracy', 'Precision', 'Recall', 'F1_score']

fig,axes=plt.subplots(3,2,figsize=(20,20))
axes=axes.ravel()
for i,(col1,col2) in enumerate(zip(data1.columns,data2.columns)):
    axes[i].plot(data1[col1],color='red',label=f'{col1}')
    axes[i].plot(data2[col2],color='blue',label=f'{col2}')
    axes[i].legend(loc='best')  # add legend for each subplot
    axes[i].set_xlabel('Epochs')  # add x label for each subplot
    axes[i].set_ylabel(labels[i]) # add y label for each subplot
    axes[i].set_title(labels[i])  # add title for each subplot
    axes[i].locator_params(axis='both',nbins=10)
axes[-1].plot(data['lr'],color='blue',label='Learning Rate')
plt.legend()
plt.title('Learning Rate')
plt.suptitle('Training and Validation Metrics',fontsize=30)
```
